```python
# method to process all data
companies_data = []
origin = 'linkedin'
for company in companies:
    
    print('Companie:', company)
    
    try:
        company_name = re.search(r'company\/(.+)\/posts', company).group(1)

        print('[Status] Processing all data from ' + company_name + '...')
        logs_file.write('[Status] Processing all data from ' + company_name + '...\n')

        with open(dir_path + company_name + '.txt', encoding="utf-8") as company_data_file:
            company_html = company_data_file.readlines()
            company_data_file.close()

        soup = BeautifulSoup(''.join(company_html), features='html.parser')
        posts_raw = soup.find_all('div', {"class": "feed-shared-update-v2"})

        
        posts_count = len(posts_raw)
        print('Total de posts:', posts_count)
        
        for i in range(posts_count):
            
            print('Post: ',i)
            
            post_title = ''
            post_date = str(datetime.date.today())
            post_text = ''
            post_likes = 0
            post_comments = 0
            post_video_views = 0
            post_hashtags = ''

            try:
                post_actor_meta = posts_raw[i].find('div', {"class": "feed-shared-actor__meta"})
                post_date_raw = post_actor_meta.find('span', {"class": "feed-shared-actor__sub-description"})
            except:
                post_date_raw = None
                print('Failures at feed-shared-actor__meta')
            
            
            post_text_raw = posts_raw[i].find('div', {"class": "feed-shared-text"})
            post_likes_raw = None
            post_comments_raw = None
            
            if posts_raw[i].find('ul', {"class": "social-details-social-counts"}) != None:
                post_likes_raw = posts_raw[i].find('ul', {"class": "social-details-social-counts"}).find('li', {"class": "social-details-social-counts__reactions"})
                post_comments_raw = posts_raw[i].find('ul', {"class": "social-details-social-counts"}).find('li', {"class": "social-details-social-counts__comments"})
                post_reactions_list_raw = posts_raw[i].find('ul', {"class": "social-details-social-counts"}).find_all('li', {"class": "social-details-social-counts__item"})
                for post_reactions_raw in post_reactions_list_raw:
                    if len(post_reactions_raw['class']) == 1:
                        post_video_views = re.search(r'([0-9\.]+)', post_reactions_raw.get_text()).group(1)
                        post_video_views = int(re.sub(r'\.', '', post_video_views))

            if post_date_raw != None:
                regex_date = re.search(r'(^[0-9]+)\ (.+?(?=\ |$))', post_date_raw.get_text(), re.MULTILINE)

                count = int(regex_date.group(1))
                unit = regex_date.group(2)

                if unit == 'min':
                    post_date = datetime.datetime.now()-datetime.timedelta(minutes=count)
                elif unit == 'h':
                    post_date = datetime.datetime.now()-datetime.timedelta(hours=count)
                elif unit == 'd':
                    post_date = datetime.datetime.now()-datetime.timedelta(days=count)
                elif unit == 'sem':
                    post_date = datetime.datetime.now()-datetime.timedelta(weeks=count)
                elif unit == 'm':
                    post_date = datetime.datetime.now()-datetime.timedelta(days=count*30)
                elif unit == 'a':
                    post_date = datetime.datetime.now()-datetime.timedelta(days=count*365)

                post_date = str(post_date.date())

            if post_text_raw != None:
                post_text = repr(post_text_raw.get_text().strip())

            if post_likes_raw != None:
                post_likes = post_likes_raw.get_text()
                post_likes = int(re.sub(r'\.', '', post_likes))

            if post_comments_raw != None:
                post_comments = re.search(r'([0-9\.]+)', post_comments_raw.get_text()).group(1)
                post_comments = int(re.sub(r'\.', '', post_comments))

            hashtags = re.findall(r'(#.+?(?=\ |$))', post_text, re.MULTILINE)
            
            if len(hashtags) != 0:
                post_hashtags = ', '.join(hashtags)

            data = {"id": i,\
                    "company": company_name,\
                    "post_title": post_title,\
                    "post_date": post_date,\
                    "post_text": post_text,\
                    "post_likes": post_likes,\
                    "post_comments": post_comments,\
                    "post_video_views": post_video_views,\
                    "post_hashtags": post_hashtags,\
                    "origin": origin}

            companies_data.append(data)

        print('[Status] Finished processing all data from '+company_name+'.')
        logs_file.write('[Status] Finished processing all data from '+company_name+'.\n')
    except:
        print('except')
```

```python
print('[Status] Writing all processed data from all companies in companies_data.csv and companies_data.parquet.')
logs_file.write('[Status] Writing all processed data from all companies in companies_data.csv and companies_data.parquet.\n')

try:
    csv_file = open(companies_posts_path)
    reader = csv.DictReader(csv_file)
    companies_data_csv = []
    for post in reader:
        companies_data_csv.append(post)
    csv_file.close()

    companies_csv = []
    companies = []
    [companies_csv.append(post['company']) for post in companies_data_csv if post['company'] not in companies_csv]
    [companies.append(post['company']) for post in companies_data if post['company'] not in companies]
    intersect_companies = list(set(companies_csv) & set(companies))
    difference_companies = list(set(companies) - set(companies_csv))

    companies_data_union = []

    for company in difference_companies:
        companies_data_union.extend([post for post in companies_data if post['company'] == company])

    for company in intersect_companies:
        company_posts_csv = [post for post in companies_data_csv if post['company'] == company]
        company_posts = [post for post in companies_data if post['company'] == company]

        company_posts_texts_csv = [post['post_text'] for post in company_posts_csv]
        company_posts_texts = [post['post_text'] for post in company_posts]
        
        company_posts_texts_intersection = list(set(company_posts_texts_csv) & set(company_posts_texts))
        
        if set(company_posts_texts_intersection) != set(company_posts_texts):
            posts = [post for post in company_posts_csv if post['post_text'] not in company_posts_texts]

            posts.extend(company_posts)

            for i in range(len(posts)):
                posts[i]['id'] = i
        else:
            posts = company_posts

        companies_data_union.extend(posts)
except:
    print('except')
    companies_data_union = companies_data
```
