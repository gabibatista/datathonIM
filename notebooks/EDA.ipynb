```python
import pandas, numpy
```

```python
df = pandas.read_csv('companies_size.csv')
df.loc[df['employees'] == 'not found', 'employees'] = numpy.nan
df.count()
```

```python
df_cli = pandas.read_csv('labeled.csv')
df_cli.head()
```

```python
df_mer = pandas.merge(
    left=df_cli,
    right=df,
    left_on='clients',
    right_on='company_searched',
    how='inner'
)

df_mer = df_mer.rename(columns={
    'company' : 'competitor',
    'clients' : 'prospects',
    'employees' : 'prospect_size'
})

labels = {
    0 : 'Platform Partners',
    1 : 'Software and Digital Transformation',
    2 : 'Software Development and Consulting',
    3 : 'Solutions for Business',
    4 : 'Labs',
    5 : 'Software Development and Consulting',
    6 : 'Data Solutions',
    7 : 'Cloud Transformation Company',
    8 : 'Data Solutions',
    9 : 'Software Development and Consulting'
}

df_mer['purchased_offer'] = df_mer['cluster'].apply(lambda x : labels.get(x))
df_mer = df_mer[[
    'competitor',
    'prospects',
    'prospect_size',
    'purchased_offer',
    'headquarters',
    'industry',
    'specialties'
]]

df_mer.head(10)
```
