```python
import os, random, sys, time, shutil, re, csv, datetime, pickle, math 

import requests
import pandas as pd
```

```python
api_key = '484afc0cd12d2557c2d0408edba9348b'
base_url = 'http://api.positionstack.com/v1/forward'
```

```python
companies_path = 'CSVs/linkedin_about/companies_about.csv'
countries_list_path = 'CSVs/commons/countries_list.csv'
```

```python
def get_companies_about():
    try: return pd.read_csv(companies_path)
    except: return pd.DataFrame()
```

```python
def get_countries_list():
    try: return pd.read_csv(countries_list_path)
    except: return pd.DataFrame()
```

```python
def get_location_for(reference):
    
    payload = {'access_key': api_key,
               'query': reference,
               'limit': 1}
    print('resquesting: ', reference)
    r = requests.get(base_url, params=payload)
    
    try: result = r.json()['data'][0]
    except: return '', '', '', ''
    
    if not result: return '', '', '', ''
    
    print('response: ', r.json()['data'])
    
    country = result.get('country', '')
    city = result.get('locality', '')
    state = result.get('region_code', '')
    code = result.get('country_code', '')
    
    return city, state, country, code
```

```python
def get_companies_headquarters_list():
    try: return get_companies_about()['headquarters'].to_list()
    except: return []
```

```python
def remove_duplicates(location_list):
    duplicated = [x for x in location_list if str(x) != 'nan']
    return list(dict.fromkeys(duplicated))
```

```python
def loadCountries(location_list):
    df_info = get_countries_list()
    for location in location_list:
        ct, st, co, cd = get_location_for(location)
        result = {"origin": location,
                  "city": ct,
                  "state": st,
                  "country": co,
                  "country_code": cd}
        df_info = df_info.append(result, ignore_index=True, verify_integrity=False)
        
    df_info.to_csv(countries_list_path, index=False)
    return df_info
```

```python
locations = get_companies_headquarters_list()
locations_filtred = remove_duplicates(locations)
locations_searched = get_countries_list()['origin'].to_list()
locations_to_search = [c for c in locations_filtred if c not in locations_searched]
loadCountries(locations_to_search)
```

```python
countries_df = get_countries_list()
companies_headquarters_list = get_companies_headquarters_list()
locations_results:[str] = []

for h in companies_headquarters_list:
    if type(h) is str:
        searched = [countries_df[countries_df['origin'] == h]][0]
        searched_country = searched['country'].values[0]
        locations_results.append(searched_country)
    else:
        locations_results.append('not found')

print(locations_results[:10])

df_result = get_companies_about()
df_result['country'] = locations_results
df_result.to_csv(companies_path, index=False)
```
